**Loading the dataset**
**Question - What are the first & last touch points of all the customers in the system.Store the data for each customer at a customer level in a separate table and call it "first_touch_point" & "last_touch_point".**




```python
import pandas as pd
df=pd.read_csv('/content/depositsdata_a2.csv')
```

**Creating one new column of first touch point**

```python
df['first_touchpoint'] = df.groupby('PLAYER_ID')['AMOUNT'].transform('first')
df['first_touchpoint2'] = df.groupby('PLAYER_ID')['WEB_MOBILE_CHANNEL'].transform('first')
```

```python
df['first_touch_point']=df['first_touchpoint2'] + "_" +  df['first_touchpoint'].astype(str)
```

**Creating one new column of last touch point**

```python
df['last_touchpoint'] = df.groupby('PLAYER_ID')['AMOUNT'].transform('last')
df['last_touchpoint2'] = df.groupby('PLAYER_ID')['WEB_MOBILE_CHANNEL'].transform('last')
```

```python
df['last_touch_point']=df['last_touchpoint2'] + "_" +  df['last_touchpoint'].astype(str)
```

```python
df
```

**Droping the individual first touch point column**

```python
df=df.drop(['first_touchpoint'],axis=1)
```

```python
df=df.drop(['first_touchpoint2'],axis=1)
df=df.drop(['last_touchpoint'],axis=1)
df=df.drop(['last_touchpoint2'],axis=1)
```

**Loading the dataset of Assignment_python**

```python
import pandas as pd
data=pd.read_csv('/content/assignment_python.csv')
```

**Calculate the time period in days for which the customer was active (days from first deposit date)** 


```python
def Customer_Lifetime(data):
  data['Customer_Lifetime'] = (pd.Timestamp.now() - pd.to_datetime(data['DEPOSIT_DATE'], dayfirst=True)).dt.days   

  return data
new_df=Customer_Lifetime(data)

```

```python
new_df
```

```python
print(new_df.loc[new_df['IDCUSTOMER'] == 5371454])
```

**Find the country with the highest number of unique customers**

```python
a=data.groupby(['COUNTRY','CITY'])['IDCUSTOMER'].nunique()
print(a)
```

**Find the city with the highest number of unique customers**

```python
df1 = data.groupby('CITY')['IDCUSTOMER'].nunique().sort_values(ascending=False).reset_index(name='count')
print (df1)
```

**Find the unique count of players in each country (in descending order) & specify the number of unique players from the city - Kayseri**

**What is the average amount?**

```python
data['AVG_DEPOSIT_AMOUNT'] = data.groupby('CITY')['FIRST_DEPOSIT_AMOUNT'].transform('mean')
print(data)
```

**Which city gives the third highest average first_deposit_amount**

```python
third_avg =data['AVG_DEPOSIT_AMOUNT'].nlargest(3)
print(third_avg)
```

```python
CITY=data.nlargest(3, ['AVG_DEPOSIT_AMOUNT'])
third_avg_city=CITY['CITY']
print(third_avg_city)
```
